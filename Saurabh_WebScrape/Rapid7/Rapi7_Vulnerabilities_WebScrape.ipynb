```python
#import libaries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from datetime import datetime
```

```python
#send request
url = "https://www.rapid7.com/db/?q=&type=nexpose"
page = requests.get(url)

#valriable to store parse datas
soup = BeautifulSoup(page.text, "html.parser")
```

```python
#read review text
review_text = []
review_text_element = soup.find_all(class_ ="resultblock__info-title")

for item in review_text_element:
    review_text.append(item.text.lstrip().rstrip())
```

```python
#read date_published
date_published = []
date_published_element = soup.find_all(class_="resultblock__info-meta")

for item in date_published_element:
    date_published.append(item.text.lstrip().rstrip())


#only_date
only_date = []
for i in date_published:
    a = i[11:30].rstrip()
    #only_date.append(a)
    datetime_object = datetime.strptime(a,'%B %d, %Y')
    only_date.append(datetime_object)
    



#only_sevirity
only_sevirity = []
for i in date_published:
    sevirity = int(i[i.find("Severity:")+9:].split()[0].lstrip())
    only_sevirity.append(sevirity)
```

```python
final_array = []
for text, date in zip(review_text,date_published):
    final_array.append({"Vulnerabilities":text,"Date_Published":date})

#create a dataframe
df = pd.DataFrame(final_array)

#adding new columns
df["Date Published"]=pd.DataFrame(only_date)
df["Severity"]=pd.DataFrame(only_sevirity)

#droping column
df = df.drop("Date_Published",axis=1)

```
